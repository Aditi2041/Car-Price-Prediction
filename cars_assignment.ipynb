## Step 3:  Data Preparation

```python
#Replacing the hyphens in the name of the cars with a space
car_details['CarName'] = car_details['CarName'].str.replace('-' , ' ')
car_details.CarName.unique()
```

```python
#Splitting the car names in the place of spaces and extracting only the first part which is the Car Company
car_details['CarName'] = car_details['CarName'].map(lambda x : x.split(' ' , 1)[0])
```

```python
car_details['CarName'] = car_details['CarName'].str.lower()
car_details['CarName'] = car_details['CarName'].str.replace('vw' , 'volkswagen')
car_details['CarName'] = car_details['CarName'].str.replace('vokswagen' , 'volkswagen')
car_details['CarName'] = car_details['CarName'].str.replace('toyouta' , 'toyota')
car_details['CarName'] = car_details['CarName'].str.replace('porcshce' , 'porsche')
car_details['CarName'] = car_details['CarName'].str.replace('maxda' , 'mazda')
car_details.CarName.unique()
```

```python
#Assigning 1 and 0 to variables which has two levels
car_details['fueltype'] = car_details['fueltype'].map({'gas' : 1 , 'diesel' : 0})
car_details['aspiration'] = car_details['aspiration'].map({'std' : 1 , 'turbo' : 0})
car_details['doornumber'] = car_details['doornumber'].map({'two' : 1 , 'four' : 0})
car_details['enginelocation'] = car_details['enginelocation'].map({'front' : 1 , 'rear' : 0})
```

## Step 4: Splitting the Data into Training and Testing Sets


```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train , df_test = train_test_split(car_details , train_size = 0.7 , test_size = 0.3 , random_state = 100)
```

### Rescaling the Features 

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
import warnings
warnings.filterwarnings('ignore')
num_vars = ['symboling' , 'wheelbase' , 'carlength' , 'carwidth' , 'carheight' , 'curbweight' , 'enginesize' , 'boreratio' , 'stroke' , 'compressionratio' , 'horsepower' , 'peakrpm' , 'citympg' , 'highwaympg' , 'price']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
df_train.head()
```

```python
df_train.describe()
```

### Dividing into X and Y sets for the model building

```python
y_train = df_train.pop('price')# so as to take the price out and assigning it to y as it is the dependent variable
x_train = df_train #assigning the rest of the variables to x as they are the independent variables
```

## Building our model

### RFE
Recursive feature elimination

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 20
lm = LinearRegression()
lm.fit(x_train , y_train)

rfe = RFE(lm , 20)
rfe = rfe.fit(x_train , y_train)
```

```python
list(zip(x_train.columns , rfe.support_ , rfe.ranking_))
```

```python
col = x_train.columns[rfe.support_]
col
```

```python
x_train.columns[~rfe.support_]
```

### Building model using statsmodel, for the detailed statistics

## Model Number 1

```python
# Creating x_train dataframe with RFE selected variables
x_train_rfe = x_train[col]
```

```python
x_train_rfe.head()# checking the newly formed dataframe
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
x_train_rfe = sm.add_constant(x_train_rfe)
```

```python
# Running the linear model
lr_1 = sm.OLS(y_train,x_train_rfe).fit()   
```

```python
#Let's see the summary of our linear model
print(lr_1.summary())
```

```python
x_train_rfe = x_train_rfe.drop(['const'], axis = 1)#dropping the constant
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = x_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
#Plotting the heatmap to check for correlation
plt.figure(figsize = (20 , 10))
sns.heatmap(x_train_rfe.corr() , annot = True , cmap = 'YlGnBu')
```

It can be seen from the heatmap that cylinder number two (two) and engine type rotor(rotor) are highly correlated with correlation 1 and the VIF is infinity for both engine type rotor and cylinder number two. Hence lets drop the feature engine type rotor and create another model
